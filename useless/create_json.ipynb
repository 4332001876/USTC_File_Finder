```python
'''
json包含以下几项内容：
  - url: 网站地址
  - encoding: 网站编码
  - title: 网站标题
  - dtype: 该列表文件类型
  - dtype2: 该列表文件类型的子类型
  - flip: 该网站是否可以翻页
  - html_locator: 该网站对应的文件列表html定位器
  - info_locator: 该网站文件列表下每个文件的信息定位器

在定位器部分，示例：
  - html_locator = [
    {'method': 'find', 'args': 'div', 'kwargs': {'class_': 'r-box'}},
    {'method': 'find', 'args': 'ul'},
    {'method': 'find_all', 'args': 'li'}
                   ]
    对应的操作是：
    bs_result = bs.find('div', class_='r-box')
    bs_result = bs_result.find('ul')
    bs_result = bs_result.find_all('li')

   - info_locator = [
    {'info': 'title', 'method': 'find', 'args': 'a', 'args2': 'text'},
    {'info': 'url', 'method': 'find', 'args': 'a', 'args2': 'href'},
    {'info': 'time', 'method': 'find', 'args': 'time', 'args2': 'text'},
                    ]
    对应的操作是：
    result[i]['title'] = bs_result[i].find('a').text.strip()
    result[i]['url'] = urljoin(url, bs_result[i].find('a')['href'])
    result[i]['time'] = bs_result[i].find('time').text.strip()
'''
```

```python
import json

def create_json(url, encoding, title, dtype, dtype2, flip, html_locator, info_locator):
    data = {
        'url': url,
        'encoding': encoding,
        'title': title,
        'dtype': dtype,
        'dtype2': dtype2,
        'flip': flip,
        'html_locator': html_locator,
        'info_locator': info_locator
    }


    data_list = []

    data_list.append(data)

    with open('data.json', 'w', encoding='utf-8') as f:
        json.dump(data_list, f, ensure_ascii=False, indent=4)

def add_data(url, encoding, title, dtype, dtype2, flip, html_locator, info_locator):
    data = {
        'url': url,
        'encoding': encoding,
        'title': title,
        'dtype': dtype,
        'dtype2': dtype2,
        'flip': flip,
        'html_locator': html_locator,
        'info_locator': info_locator
    }


    with open('data.json', 'r') as json_file:
        data_list = json.load(json_file)

    data_list.append(data)

    with open('data.json', 'w', encoding='utf-8') as f:
        json.dump(data_list, f, ensure_ascii=False, indent=4)
```
